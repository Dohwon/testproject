```python
import pandas as pd
```

```python
data = pd.read_csv('validation.csv', encoding = 'CP949',names=['intent','sent'])
ask = pd.read_csv('QUESTION.csv', encoding = 'CP949',names=['sent'])
ask1 = pd.read_csv('QUESTION.csv', encoding = 'CP949',names=['sent'])
```

```python
ask = pd.merge(ask, data, how='left',
        left_on='sent', right_on='sent')
ask = ask.drop_duplicates(['sent'])
ask = ask.reset_index(drop=True)
```

```python
ask1 = pd.merge(ask,ask1, how='left', left_on = 'sent', right_on = 'sent')
ask1
```
