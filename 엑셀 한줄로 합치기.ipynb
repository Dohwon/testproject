```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('도원씨송부파일.csv', encoding = 'UTF-8')
```

```python
datare = data.iloc[:,3:]
```

```python
datare = datare.T
```

```python
datare = datare.reset_index(drop=True , inplace=True)
datare
```

```python
file = pd.DataFrame(columns=['UserSentences'])
```

```python
for i in range(138):
    newfile = pd.DataFrame(columns=['UserSentences'])
    newfile['UserSentences'] = datare.iloc[:,i].dropna()
    file = pd.concat([file,newfile])
```

```python
file.to_csv('정리본.csv')
```
