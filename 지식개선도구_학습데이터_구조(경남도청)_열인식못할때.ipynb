```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
user = pd.read_csv('intent_userSentence.csv',encoding = "utf-8",names = list(range(0,500)))
user.head()
```

```python
df = pd.DataFrame(user)
df = df.drop([1,2], axis=1)
df
```

```python
df1 = pd.melt(df, id_vars=[0], value_vars=None)   #id_vars=['대분류','중분류','소분류','인텐트명']
df1.head(10)
```

```python
df1.dropna(inplace=True)                # 공백이 들어있는 행삭제
del df1['variable']                     # 필요없는 열 삭제
df1 = df1.sort_values(by=[0], axis=0)  # 오름차순
```
